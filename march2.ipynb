## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

####  In a scatter plot, each data point is represented as a point on the plot with its x and y values representing the values of the two variables being compared.

```python
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Demo Plot')
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

#### The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure. It allows us to arrange multiple plots in a grid-like structure, with each plot sharing the same figure and axis.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axs[0, 0].plot(x1,y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 1].plot(x3, y3)
axs[1, 1].set_title('Line 3')

axs[1, 0].plot(x4, y4)
axs[1, 0].set_title('Line 4')
fig.suptitle('Four Line Plots')
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

The box plot is a type of plot used to display the distribution of a set of continuous data. 
The box plot consists of a box with whiskers extending from the top and bottom edges, and possibly also outliers plotted as individual points.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
plt.show()
```
