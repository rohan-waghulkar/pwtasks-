## Q3. What is App routing in Flask? Why do we use app routes?

Routing in flask is a proccess of mapping URL to python functions that handles those URLs.

We use app routing in Flask to define the URLs that a web application should respond to, as well as the Python functions that generate the response for each URL.

to achive app routing  @app.route() is used.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function is used in Flask for URL Building.

from flask import Flask, redirect,url_for


app = Flask(__name__)

@app.route('/lesser/<int:num>')

def lesser(num):
    return "the number is less than or equals to 5 "

@app.route("/greater/<int:num>")

def greater(num):
return "the num is greater than 5 "

@app.route("/getnum/<int:number>",)

def getnum(number):
    if number>5:
        result="greater"
    else :
        result="lesser"
    
    return redirect(url_for(result,num=number))

if __name__=="__main__":
    app.run(host="0.0.0.0")
