## Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

- Fixed acidity: The amount of non-volatile acids in the wine, which gives it a tartness.
- Volatile acidity: The amount of acetic acid in the wine, which gives it a vinegary taste.
- Citric acid: A type of acid that gives wine a tartness and freshness.
- Residual sugar: The amount of sugar remaining in the wine after fermentation.
- Chlorides: The amount of chloride ions in the wine, which can give it a salty taste.
- Free sulfur dioxide: The amount of free sulfur dioxide in the wine, which is used as a preservative.
- Total sulfur dioxide: The total amount of sulfur dioxide in the wine, including both free and bound sulfur dioxide.
- Density: The density of the wine, which is affected by the alcohol content and other factors.
- pH: The acidity of the wine, which is measured on a scale of 0 to 14.
- Sulphates: The amount of sulfates in the wine, which can give it a bitter taste.
- Alcohol: The percentage of alcohol in the wine.


- Of these features, the most important ones in predicting wine quality are alcohol, volatile acidity, and sulphates. Wines with higher levels of alcohol and lower levels of volatile acidity and sulphates tend to be rated higher in quality

#### Imputation Techiques

- Mean imputation replaces missing values with the mean value of the corresponding feature. This is the most common imputation technique, and it is often a good choice when the data is normally distributed. However, mean imputation can lead to overfitting, and it can underestimate the variance of the data.

- Mode imputation replaces missing values with the mode value of the corresponding feature. This is a less biased imputation technique than mean imputation, but it can lead to underestimation of the variance of the data.

- Median imputation replaces missing values with the median value of the corresponding feature. This is a robust imputation technique that is not affected by outliers. However, median imputation can lead to underestimation of the mean of the data.

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Separate the features from the target variable (quality)
X = df.drop('quality', axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Calculate cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Determine the minimum number of principal components required for 90% variance
n_components = np.argmax(cumulative_variance_ratio >= 0.9) + 1

print("Minimum number of principal components to explain 90% of variance:", n_components)
```
