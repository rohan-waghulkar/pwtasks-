## Q2. Why is request used in Flask?

The request object allows the Flask application to access the incoming request data and to interact with the client.

the request object is used to handle incoming requests from clients. When a client sends a request to the Flask application, the request object is created and contains all the relevant information about the request, such as the headers, data, and URL.
