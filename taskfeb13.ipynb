## Q1. Explain why we have to use the Exception class while creating a Custom Exception.

it is important to use the built-in Exception class as the base class for your custom exception because it provides important functionality that is necessary for proper exception handling.
 such as :
        
        1: in python all exceptions are derived from BaseException class . The Exception class is a subclass of BaseException, it inherits the all important properties of BaseException class which are important for Exception handling.
        
        2: The Exception class provides constructors that allow you to set a message and optionally, additional arguments. This information is important for debugging and error reporting.
        
        3: The Exception class provides a default constructor, which is important for creating instances of the exception.
        
        4: By using the Exception class as the base class for your custom exception, you ensure that your exception is consistent with other exceptions in the Python language and provides the necessary functionality for proper exception handling.

## Q2. Write a python program to print Python Exception Hierarchy.

def printclass(cname,indent=0):
    print(" "*indent,"|--",cname.__name__)
    for subclass in cname.__subclasses__():
        printclass(subclass,indent+2)

```python
printclass(Exception)
```

## Q5. Explain ImportError. What is ModuleNotFoundError?

 ImportError: is a built-in exception that is raised when a module or package cannot be imported. This can occur for a number of reasons, such as a missing or invalid module name, an invalid path to the module, or a syntax error in the module itself.
 The ModuleNotFoundError exception is raised when the interpreter is unable to locate the module that is being imported. This can occur for a number of reasons, such as when the module is not installed, or when the module is installed in a location that is not in the Python path.
